```python
from PIL import Image, ImageDraw, ImageFont, ImageOps, ImageEnhance
```

```python
size = (512,512)
img = Image.new("L", size, (0,))
```

```python
atkinson = ImageFont.truetype("Atkinson-Hyperlegible-Regular-102.otf",100)
```

```python
draw = ImageDraw.Draw(img)
```

```python
draw.text((0,0), "22\n59", (255,), font=atkinson)
```

```python
img
```

```python
cropped = img.crop(img.getbbox())
cropped
```

```python
cropped.size
```

```python
max(*cropped.size)
```

```python
ImageOps.pad(cropped, (max(*cropped.size), max(*cropped.size)))
```

```python
resized = ImageOps.expand(ImageOps.pad(cropped, (max(*cropped.size), max(*cropped.size))), 1).resize(size, resample=Image.Resampling.LANCZOS)
resized
```

```python
enhanced = ImageEnhance.Contrast(resized).enhance(9000)
enhanced
```

```python
enhanced.save("2259.png")
```
